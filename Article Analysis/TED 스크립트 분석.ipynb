```python
import re
import urllib.request
import zipfile
from lxml import etree
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/ukairia777/tensorflow-nlp-tutorial/main/09.%20Word%20Embedding/dataset/ted_en-20160408.xml", filename="files/ted_en-20160408.xml")
```

```python
with open('ted_en-20160408.xml', 'r', encoding='UTF8') as targetXML:
     target_text = etree.parse(targetXML)

# xml  <content> </content>   .
parse_text = '\n'.join(target_text.xpath('//content/text()'))

#   sub   content   (Audio), (Laughter)    .
#      .
content_text = re.sub(r'\([^)]*\)', '', parse_text)

#  
content_text = content_text[:len(content_text) // 10]

#    NLTK    .
sent_text = sent_tokenize(content_text)

#     ,   .
normalized_text = []
for string in sent_text:
     tokens = re.sub(r"[^a-z0-9]+", " ", string.lower())
     normalized_text.append(tokens)

#    NLTK    .
result = [word_tokenize(sentence) for sentence in normalized_text]
result
```
